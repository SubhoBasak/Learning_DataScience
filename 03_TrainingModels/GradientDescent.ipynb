```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2*np.random.rand(100, 1)
y = 10+3*X+np.random.rand(100, 1)
```

```python
plt.scatter(X, y)
```

```python
theta_0 = theta_1 = 0
lrn_rate = 0.001

line_func = lambda X: theta_0+theta_1*X
```

```python
def plot_line(func, X_points):
    y_points = [func(i) for i in X_points]
    plt.plot(X_points, y_points)
```

```python
plot_line(line_func, X)
```

```python
def summetion(func, X_points, y_points):
    total1 = 0
    total2 = 0
    m = len(X_points)
    
    for i in range(1, m):
        total1 += func(X_points[i]) - y_points[i]
        total2 += total1 * X_points[i]
    
    return total1/m, total2/m
```

```python
n_iterations = 1000

s1, s2 = summetion(line_func, X, y)

for i in range(n_iterations):
    theta_0 = theta_0 - lrn_rate * s2
    theta_1 = theta_1 - lrn_rate * s1
    
    if i%100 == 0:
        plot_line(line_func, X)

plot_line(line_func, X)
plt.scatter(X, y)
plt.show()
```
