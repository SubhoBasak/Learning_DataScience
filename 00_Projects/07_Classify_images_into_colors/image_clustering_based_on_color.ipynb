```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from PIL import Image
```

```python
data = []

for img_path in os.listdir('Images'):
    img_path = os.path.join('Images', img_path)
    img = Image.open(img_path)
    data.append(np.array(img.resize((50, 50))).ravel())
```

```python
plt.figure(figsize = (10, 3))
plt.subplot(1, 3, 1)
plt.imshow(data[5].reshape(50, 50, 3))
plt.subplot(1, 3, 2)
plt.imshow(data[10].reshape(50, 50, 3))
plt.subplot(1, 3, 3)
plt.imshow(data[32].reshape(50, 50, 3))
```

```python
for inst in data:
    kmeans = KMeans(n_clusters = 5)
    kmeans.fit(inst.reshape(-1, 3))
    
    img_segment = kmeans.cluster_centers_[kmeans.labels_].astype('int32')
    plt.imshow(img_segment.reshape(50, 50, 3))
    plt.show()
```
