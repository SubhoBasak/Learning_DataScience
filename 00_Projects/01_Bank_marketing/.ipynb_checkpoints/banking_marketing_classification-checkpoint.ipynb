```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
df = pd.read_csv('bank-additional-full.csv', delimiter = ';')
```

```python
df.info()
```

```python
class MonthEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = np.array(X.ravel())
        for i in range(len(X)):
            m = X[i];
            if X[i] == 'jan':
                X[i] = 1
            elif X[i] == 'feb':
                X[i] = 2
            elif X[i] == 'mar':
                X[i] = 3
            elif X[i] == 'apr':
                X[i] = 4
            elif X[i] == 'may':
                X[i] = 5
            elif X[i] == 'jun':
                X[i] = 6
            elif X[i] == 'jul':
                X[i] = 7
            elif X[i] == 'aug':
                X[i] = 8
            elif X[i] == 'sep':
                X[i] = 9
            elif X[i] == 'oct':
                X[i] = 10
            elif X[i] == 'nov':
                X[i] = 11
            elif X[i] == 'dec':
                X[i] = 12
        return X.astype('int32').reshape(-1, 1)
```

```python
class DayEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = np.array(X.ravel())
        for i in range(len(X)):
            if X[i] == 'sun':
                X[i] = 0
            elif X[i] == 'mon':
                X[i] = 1
            elif X[i] == 'tue':
                X[i] = 2
            elif X[i] == 'wed':
                X[i] = 3
            elif X[i] == 'thu':
                X[i] = 4
            elif X[i] == 'fri':
                X[i] = 5
            elif X[i] == 'sat':
                X[i] = 6

        return X.astype('int32').reshape(-1, 1)
```

```python
class EducationEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = np.array(X.ravel())
        for i in range(len(X)):
            if X[i] == 'basic.4y':
                X[i] = 1
            elif X[i] == 'basic.6y':
                X[i] = 2
            elif X[i] == 'basic.9y':
                X[i] = 3
            elif X[i] == 'high.school':
                X[i] = 4
            elif X[i] == 'illiterate':
                X[i] = 5
            elif X[i] == 'professional.course':
                X[i] = 6
            elif X[i] == 'university.degree':
                X[i] = 7
            elif X[i] == 'unknown':
                X[i] = 2
        
        return X.astype('int32').reshape(-1, 1)
```

```python
df['y'].replace('yes', 1, inplace = True)
df['y'].replace('no', 0, inplace = True)
```

```python
month_enc = MonthEncoder()
df['month'] = month_enc.fit_transform(df['month'])
```

```python
day_enc = DayEncoder()
df['day_of_week'] = day_enc.fit_transform(df['day_of_week'])
```

```python
education_enc = EducationEncoder()
df['education'] = education_enc.fit_transform(df['education'])
```

```python
class ThreeWayEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = np.array(X.ravel())
        for i in range(len(X)):
            if X[i] == 'no' or X[i] == 'failure':
                X[i] = 0
            elif X[i] == 'yes' or X[i] == 'success':
                X[i] = 1
            elif X[i] == 'unknown' or X[i] == 'nonexistent':
                X[i] = -1
        return X.reshape(-1, 1).astype('int32')
```

```python
default_enc = ThreeWayEncoder()
housing_enc = ThreeWayEncoder()
loan_enc = ThreeWayEncoder()
poutcome_enc = ThreeWayEncoder()

df['default'] = default_enc.fit_transform(df['default'])
df['housing'] = housing_enc.fit_transform(df['housing'])
df['loan'] = loan_enc.fit_transform(df['loan'])
df['poutcome'] = poutcome_enc.fit_transform(df['poutcome'])
```

```python
# ================== job ==================
job_enc = OneHotEncoder()
job_cat = job_enc.fit_transform(df['job'].values.reshape(-1, 1)).toarray().astype('int32')

job_title = np.array(df['job'].value_counts().index)
job_title.sort()
for i, j in enumerate(job_title):
    df[j] = job_cat[:, i]

# ================== marital ==================
marital_enc = OneHotEncoder()
marital_cat = marital_enc.fit_transform(df[['marital']]).toarray().astype('int32')

marital_title = np.array(df['marital'].value_counts().index)
marital_title.sort()
for i, j in enumerate(marital_title):
    df[j] = marital_cat[:, i]

# ================== contact ==================
contact_enc = OneHotEncoder()
contact_cat = contact_enc.fit_transform(df[['contact']]).toarray().astype('int32')

contact_title = np.array(df['contact'].value_counts().index)
contact_title.sort()
for i, j in enumerate(contact_title):
    df[j] = contact_cat[:, i]
```

```python
df.drop('marital', axis = 1, inplace = True)
df.drop('contact', axis = 1, inplace = True)
df.drop('job', axis = 1, inplace = True)
```

```python
X = df.drop('y', axis = 1)
y = df['y'].copy()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

```python
score_board = {'accuracy': [], 'recall': [], 'precision': [], 'f1': []}
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(max_iter = 2000)
log_reg.fit(X_train, y_train)

log_reg.score(X_test, y_test)
```

```python
y_pred = log_reg.predict(X_test)

score = accuracy_score(y_test, y_pred)
score_board['accuracy'].append(score)

score = recall_score(y_test, y_pred)
score_board['recall'].append(score)

score = precision_score(y_test, y_pred)
score_board['precision'].append(score)

score = f1_score(y_test, y_pred)
score_board['f1'].append(score)
```
