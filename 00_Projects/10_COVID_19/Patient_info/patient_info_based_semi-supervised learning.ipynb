```python
from sklearn.preprocessing import LabelEncoder
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class MLabelEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.le = LabelEncoder()
        
    def fit(self, X):
        X = np.array(X)
        self.le.fit(X)
        return self
    
    def transform(self, X):
        X = np.array(X)
        tmp_lst = []
        for i in X:
            if i in self.le.classes_:
                tmp_lst.append(self.le.transform([i])[0])
            else:
                tmp_lst.append(-1)
        return np.array(tmp_lst).reshape(-1, 1)
```

```python
def object_median(X, ignore_values = []):
    med = None
    mx = 0
    X = list(X)
    uniq = np.unique(X)
    for i in uniq:
        if i in ignore_values:
            continue
        c = X.count(i)
        if c > mx:
            mx = c
            med = i

    return med
```

```python
def estimate_age(X):
    X = np.array(X)
    for i in range(len(X)):
        X[i] = int(X[i].strip('s'))
    return X
```

```python
state_prob_based_on_age = {}

for age in np.unique(train_set['age']):
    print('\nAge group : ', age)
    print('==========================')
    dec = len(train_set[(train_set['age'] == age) & (train_set['state'] == 'deceased')])
    iso = len(train_set[(train_set['age'] == age) & (train_set['state'] == 'isolated')])
    rel = len(train_set[(train_set['age'] == age) & (train_set['state'] == 'released')])
    sm = dec+iso+rel
    dec /= sm
    iso /= sm
    rel /= sm
    print('Deceased : ', round(dec, 3))
    print('Isolated : ', round(iso, 3))
    print('Released : ', round(rel, 3))
    if age not in state_prob_based_on_age.keys():
        state_prob_based_on_age[age] = []
    state_prob_based_on_age[age].append([dec, iso, rel])
```

```python
def prepared_test_data(X):
    X.drop('contact_number', axis = 1, inplace = True)
    X.drop('patient_id', axis = 1, inplace = True)
    X.drop('country', axis = 1, inplace = True)
    X.drop('infection_order', axis = 1, inplace = True)
    X.drop(['infection_case', 'infected_by'], axis = 1, inplace = True)
    X.drop(['symptom_onset_date', 'confirmed_date', 'released_date', 'deceased_date'], axis = 1, inplace = True)
    X.drop(['city', 'province'], axis = 1, inplace = True)
    X.drop('global_num', axis = 1, inplace = True)
    X.drop('sex', axis = 1, inplace = True)

    X.fillna({'age': age_median}, inplace = True)
    X['age'] = estimate_age(X['age'])
    X['age'] = X['age'].astype('int32')
    
    tmp_list = []
    for i in range(len(X)):
        if pd.isnull(X['disease'].loc[i]):
            if X['age'].loc[i] < 30:
                tmp_list.append(0)
            elif X['state'].loc[i] == 'released':
                tmp_list.append(0)
            elif X['state'].loc[i] == 'deceased':
                tmp_list.append(1)
            else:
                tmp_list.append(0.5)
        else:
            tmp_list.append(1)
    X['disease'] = tmp_list

    X.fillna({'birth_year': train_set['birth_year'].median()}, inplace = True)
    
    X['state'] = state_enc.transform(X['state'])
    return X
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict, GridSearchCV
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import precision_recall_curve, roc_curve
from IPython.display import display, Markdown
from sklearn.preprocessing import OneHotEncoder
```

```python
def show_all_scores(y1, y2):
    display(Markdown('## Accuracy score : '+str(accuracy_score(y1, y2))))
    display(Markdown('## Recall score : '+str(recall_score(y1, y2, average = 'weighted'))))
    display(Markdown('## Precision score : '+str(precision_score(y1, y2, average = 'weighted'))))
    display(Markdown('## F1 score : '+str(f1_score(y1, y2, average = 'weighted', labels = np.unique(y2)))))

def plot_all_curves(y1, y2):
    plt.figure(figsize = (10, 15))
    
    plt.subplot(3, 1, 1)
    plt.title('preicison recall curve with threshold', fontsize = 12, color = 'orange')
    for i in range(3):
        precision, recall, threshold = precision_recall_curve(y1[:, i], y2[:, i])
        plt.plot(threshold, precision[:-1], label = 'precision')
        plt.plot(threshold, recall[:-1], label = 'recall')
    plt.grid(True)
    plt.legend()
    plt.xlabel('threshold', fontsize = 11, color = 'blue')
    plt.ylabel('score', fontsize = 11, color = 'blue', rotation = 90)
    
    plt.subplot(3, 2, 3)
    plt.title('precision recall curve', fontsize = 12, color = 'orange')
    plt.plot([0, 1], 'w--')
    for i in range(3):
        precision, recall, _ = precision_recall_curve(y1[:, i], y2[:, i])
        plt.plot(recall, precision)
    plt.grid()
    plt.xlabel('recall', fontsize = 11, color = 'blue')
    plt.ylabel('precision', fontsize = 11, color = 'blue', rotation = 90)
    
    plt.subplot(3, 2, 4)
    plt.title('roc curve', fontsize = 12, color = 'orange')
    plt.plot([0, 1], 'w--')
    for i in range(3):
        fpr, tpr, _ = roc_curve(y1[:, i], y2[:, i])
        plt.plot(fpr, tpr)
    plt.grid(True)
    plt.xlabel('false positive rate', fontsize = 11, color = 'blue')
    plt.ylabel('true positive rate', fontsize = 11, color = 'blue')
    plt.title('roc curve with threshold', fontsize = 12, color = 'orange')
    
    plt.subplot(3, 1, 3)
    for i in range(3):
        fpr, tpr, threshold = roc_curve(y1[:, i], y2[:, i])
        plt.plot(threshold, fpr, label = 'FPR')
        plt.plot(threshold, tpr, label = 'TPR')
    plt.grid(True)
    plt.legend()
    plt.xlabel('threshold', fontsize = 11, color = 'blue')
    plt.ylabel('score', fontsize = 11, color = 'blue')
    
    
    plt.show()
```
