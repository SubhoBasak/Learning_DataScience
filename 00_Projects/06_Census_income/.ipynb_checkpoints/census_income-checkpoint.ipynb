```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
df['education'].value_counts().sort_index().index
```

```python
class EducationEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X):
        return self
    
    def transform(self, X):
        result = []
        X = np.array(X).ravel()
        for i in X:
            if i == ' Preschool':
                result.append(0)
            elif i == ' 1st-4th':
                result.append(1)
            elif i == ' 5th-6th':
                result.append(2)
            elif i == ' 7th-8th':
                result.append(3)
            elif i == ' 9th':
                result.append(4)
            elif i == ' 10th':
                result.append(5)
            elif i == ' 11th':
                result.append(6)
            elif i == ' 12th':
                result.append(7)
            elif i == ' HS-grad':
                result.append(8)
            elif i == ' Assoc-acdm':
                result.append(9)
            elif i == ' Assoc-voc':
                result.append(10)
            elif i == ' Some-college':
                result.append(11)
            elif i == ' Bachelors':
                result.append(12)
            elif i == ' Prof-school':
                result.append(13)
            elif i == ' Masters':
                result.append(14)
            elif i == ' Doctorate':
                result.append(15)
        return np.array(result).reshape(-1, 1)
```

```python
education_enc = EducationEncoder()
df['education'] = education_enc.fit_transform(df['education'])
```

```python
df = pd.read_csv('adult.data.csv', names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'],
                 na_values = ' ?', index_col = False)
df.head(5)
```

```python
median_list = []

for col in list(df):
    if df[col].dtype != 'O':
        continue
    median_list.append(df[col].value_counts().sort_values(ascending = True).index[0])

median_list
```

```python
count = 0
for col in list(df):
    if df[col].dtype == 'O':
        df[col].fillna(median_list[count], inplace = True)
        count += 1
    else:
        med = df[col].median()
        df[col].fillna(med, inplace = True)
```

```python
df.info()
```

```python
for col in list(df):
    if df[col].dtype != 'O':
        continue
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
df.info()
```

```python
cor_mat = df.corr()
x = cor_mat['income'].sort_values(ascending = False).values
```

```python
plt.hist(x)
```

```python
sum(x)/len(x)
```

```python
X = df.drop('income', axis = 1)
y = df['income'].copy()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
splitter = StratifiedShuffleSplit(test_size = 0.3, n_splits = 5, random_state = 33)

for train_indx, test_indx in splitter.split(X, y):
    strat_train_x = X.loc[train_indx]
    strat_train_y = y.loc[train_indx]
    strat_test_x = X.loc[test_indx]
    strat_test_y = y.loc[test_indx]
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_curve, precision_recall_curve, f1_score
from sklearn.model_selection import cross_val_score, cross_val_predict, GridSearchCV
```

## SGDClassifier

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier()
sgd_clf.fit(strat_train_x, strat_train_y)
```

```python
sgd_clf.score(strat_test_x, strat_test_y)
```

```python
sgd_clf = SGDClassifier()

params = {
    'alpha': [0.001, 0.005, 0.0001, 0.0005, 0.00001],
    'l1_ratio': [0.1, 0.15, 0.2, 0.25, 0.3]
}

grid_model = GridSearchCV(sgd_clf, param_grid = params, cv = 3)
grid_model.fit(strat_train_x, strat_train_y)
```

```python
grid_model.score(strat_test_x, strat_test_y)
```

```python
grid_model.best_params_
```

```python
sgd_clf = SGDClassifier(**grid_model.best_params_)
sgd_clf.fit(strat_train_x, strat_train_y)
```

```python
y_pred = sgd_clf.predict(strat_test_x)
```

```python
accuracy_score(strat_test_y, y_pred)
```

```python
recall_score(strat_test_y, y_pred)
```

```python
precision_score(strat_test_y, y_pred)
```

```python
f1_score(strat_test_y, y_pred)
```

```python
sgd_clf = SGDClassifier(**grid_model.best_params_)
scores = cross_val_score(sgd_clf, X, y)

print(scores)
print(scores.mean())
```

```python
y_pred = cross_val_predict(sgd_clf, strat_train_x, strat_train_y, cv = 3, method = 'decision_function')
```

```python
precision, recall, threshold = precision_recall_curve(strat_train_y, y_pred)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(threshold, precision[:-1], 'g', label = 'precision')
plt.plot(threshold, recall[:-1], 'b', label = 'recall')
plt.legend()
plt.grid(True)
plt.xlabel('Threshold', fontsize = 12)
plt.ylabel('Score', fontsize = 12, rotation = 0)
```

```python
plt.plot(recall, precision)
plt.xlabel('Recall', fontsize = 12)
plt.ylabel('Precision', fontsize = 12, rotation = 90)
plt.grid(True)
```

```python
fpr, tpr, threshold = roc_curve(strat_train_y, y_pred)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(threshold, tpr, label = 'True positive rate')
plt.plot(threshold, fpr, label = 'Flase positive rate')
plt.legend()
plt.grid()
plt.xlabel('Threshold', fontsize = 12)
plt.ylabel('Score', fontsize = 12, rotation = 90)
```

```python
plt.plot(fpr, tpr)
plt.plot([0, 1], 'k--')
plt.grid()
plt.xlabel('False positive rate', fontsize = 12)
plt.ylabel('True positive rate', fontsize = 12, rotation = 90)
```
