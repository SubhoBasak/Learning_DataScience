```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
X = df.drop('edibility', axis = 1)
y = df['edibility'].copy()
```

```python
splitter = StratifiedShuffleSplit(test_size = 0.3, n_splits = 5)

for train_index, test_index in splitter.split(X, y):
    strat_train_x = X.loc[train_index]
    strat_train_y = y.loc[train_index]
    strat_test_x = X.loc[test_index]
    strat_test_y = y.loc[test_index]
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, precision_recall_curve, roc_curve
from sklearn.model_selection import cross_val_score, cross_val_predict, GridSearchCV
from IPython.display import display, Markdown
```

```python
def show_all_scores(y1, y2):
    display(Markdown('### Accuracy : '+str(round(accuracy_score(y1, y2), 4))))
    display(Markdown('### Recall : '+str(round(recall_score(y1, y2), 4))))
    display(Markdown('### Precision : '+str(round(precision_score(y1, y2), 4))))
    display(Markdown('### F1 score : '+str(round(f1_score(y1, y2), 4))))

def plot_precision_recall_curve_with_threshold(precision, recall, threshold):
    plt.figure(figsize = (10, 4))
    plt.title('Precision Recall curve with threshold')
    plt.plot(threshold, recall[: -1], 'g--', label = 'Recall')
    plt.plot(threshold, precision[: -1], 'b-', label = 'Precisison')
    plt.legend()
    plt.grid()
    plt.xlabel('Threshold', fontsize = 15, rotation = 0)
    plt.ylabel('Score', fontsize = 15, rotation = 90)

def plot_precision_recall_curve(precision, recall):
    plt.figure(figsize = (10, 7))
    plt.title('Precision Recall curve')
    plt.plot(recall, precision)
    plt.grid()
    plt.xlabel('Recall', fontsize = 15, rotation = 0)
    plt.ylabel('Precision', fontsize = 15, rotation = 90)

def plot_roc_curve_with_threshold(fpr, tpr, threshold):
    plt.figure(figsize = (10, 4))
    plt.title('ROC curve with threshold')
    plt.plot(threshold, fpr, 'g--', label = 'False positive rate')
    plt.plot(threshold, tpr, 'b-', label = 'True positive rate')
    plt.legend()
    plt.grid()
    plt.xlabel('Threshold', fontsize =  15, rotation = 0)
    plt.ylabel('Score', fontsize = 15, rotation = 90)

def plot_roc_curve(fpr, tpr):
    plt.figure(figsize = (10, 7))
    plt.title('ROC curve')
    plt.plot(fpr, tpr)
    plt.grid()
    plt.xlabel('False positive rate', fontsize = 15, rotation = 0)
    plt.ylabel('True positive rate', fontsize = 15, rotation = 90)
```

```python
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier
```

```python
param_list = [
    {'penalty': ['l1', 'l2'], 'alpha': [0.1, 0.01, 0.001, 0.0001], 'l1_ratio': [0.15, 0.30, 0.60, 0.75], 'max_iter': [500, 1000]},  # SGDClassifier
    {'penalty': ['l1', 'l2'], 'tol': [0.001, 0.0001, 0.00001], 'C': [5, 10, 15], 'max_iter': [30000]}, #LogisticRegression
    {'C': [5, 10, 15], 'tol': [0.001, 0.0001, 0.00001], 'degree': [1, 2, 3, 4], 'probability': [True]}, #SVC
    {'max_leaf_nodes': [10, 20, 30, 40], 'max_depth': [7, 9, 10, 15], 'splitter': ['best', 'random']}, #DecisionTreeClassifier
    {'n_estimators': [50, 100, 200], 'max_depth': [7, 9, 10, 15], 'max_leaf_nodes': [10, 20, 30, 40]}, #RandomForestClassifier
    {'max_depth': [7, 9, 10, 12], 'max_leaf_nodes': [7, 9, 12]}, #ExtraTreesClassifier
    {'n_estimators': [50, 70, 100, 200], 'learning_rate': [1.0, 0.1, 0.001]}, #AdaBoostClassifier
    {'n_estimators': [60, 70, 100, 150], 'learning_rate': [0.1, 0.01, 0.001]} #GradientBoostingClassifier
]
```

```python
clfs = [SGDClassifier, LogisticRegression, SVC, DecisionTreeClassifier, RandomForestClassifier, ExtraTreesClassifier,
        AdaBoostClassifier, GradientBoostingClassifier]
best_clfs = []
best_params = []
score_board = {'Classifier': [], 'Accuracy': [], 'Recall': [], 'Precision': [], 'F1 score': []}
```

```python
# strat_train_x = strat_train_x[: 50]
# strat_train_y = strat_train_y[: 50]
# strat_test_x = strat_test_x[: 50]
# strat_test_y = strat_test_y[: 50]

for cur_clf, cur_param in zip(clfs, param_list):
    clf = cur_clf()
    display(Markdown('## '+str(clf.__class__.__name__)))
    clf.fit(strat_train_x, strat_train_y)
    
    print('\n\nScore with default params')
    print('=========================')
    print('score : ', clf.score(strat_test_x, strat_test_y))
    
    clf = cur_clf()
    grid_model = GridSearchCV(clf, param_grid = cur_param, cv = 3)
    grid_model.fit(strat_train_x, strat_train_y)
    
    print('\n\nBest params after grid search')
    print('=============================')
    print(grid_model.best_params_)
    
    clf = cur_clf(**grid_model.best_params_)
    scores = cross_val_score(clf, strat_train_x, strat_train_y, cv = 5)
    
    print('\n\nCross val score after grid search')
    print('=================================')
    print(scores)
    print(scores.mean())
    
    clf.fit(strat_train_x, strat_train_y)
    y_pred = clf.predict(strat_test_x)
    
    
    show_all_scores(strat_test_y, y_pred)
    
    clf = cur_clf(**grid_model.best_params_)
    if clf.__class__.__name__ == 'SGDClassifier':
        y_pred = cross_val_predict(clf, strat_train_x, strat_train_y, cv = 3, method = 'decision_function')
    else:
        y_pred = cross_val_predict(clf, strat_train_x, strat_train_y, cv = 3, method = 'predict_proba')[:, -1]
    
    precision, recall, threshold = precision_recall_curve(strat_train_y, y_pred)
    plot_precision_recall_curve_with_threshold(precision, recall, threshold)
    plt.show()
    plot_precision_recall_curve(precision, recall)
    plt.show()
    
    fpr, tpr, threshold = roc_curve(strat_train_y, y_pred)
    plot_roc_curve_with_threshold(fpr, tpr, threshold)
    plt.show()
    plot_roc_curve(fpr, tpr)
    plt.show()
```
