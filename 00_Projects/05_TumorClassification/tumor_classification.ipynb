```python
X = df.drop('class', axis = 1)
y = df['class'].copy()
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3)

spliter = StratifiedShuffleSplit(n_splits = 3, test_size = 0.3)

for train_index, test_index in spliter.split(X, y):
    strat_train_x = X.loc[train_index]
    strat_train_y = y.loc[train_index]
    strat_test_x = X.loc[test_index]
    strat_test_y = y.loc[test_index]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV, cross_val_predict, cross_val_score
```

```python
score_board = {'Classifier': [], 'Accuracy': []}
grid_best_params = {'Classifier': [], 'Params': []}
```

```python
def show_params():
    for i, j in zip(grid_best_params['Classifier'], grid_best_params['Params']):
        print(i, j)
```

```python
# log_reg = LogisticRegression()
# svm_clf = SVC(probability = True)
# tree_clf = DecisionTreeClassifier()
# forest_clf = RandomForestClassifier()
# extra_clf = ExtraTreesClassifier()

# clf_list = [LogisticRegression, SVC, DecisionTreeClassifier, RandomForestClassifier, ExtraTreesClassifier]
clf_list = [SVC, DecisionTreeClassifier, RandomForestClassifier, ExtraTreesClassifier]
```

```python
param_list = [
#     {'C': [0.5, 1.0, 1.5, 2.0], 'max_iter': [10000]},
    {'C': [8, 10, 12, 15], 'degree': [2, 3, 4]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45], 'max_features': [15, 18, 20]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45], 'max_features': [15, 18, 20]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45], 'max_features': [15, 18, 20]},
]
```

```python
for clf, params in zip(clf_list, param_list):
    model = clf()
    grid_model = GridSearchCV(model, param_grid = params, cv = 3)
    grid_model.fit(X, y)
    grid_best_params['Classifier'].append(model.__class__.__name__)
    grid_best_params['Params'].append(grid_model.best_params_)
    
    model = clf(**grid_model.best_params_)
    score_board['Classifier'].append(model.__class__.__name__)
    score_board['Accuracy'].append(cross_val_score(model, X, y, cv = 3))

    print(model.__class__.__name__, ':: checking complete')
```

```python
show_params()
```

```python
score_board_df = pd.DataFrame(score_board)
score_board_df
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 20)
data = pca.fit_transform(X)
```

```python
score_board = {'Classifier': [], 'Accuracy': [], 'Accuracy mean': [], 'N_features': []}
grid_best_params = {'Classifier': [], 'Params': []}
```

```python
param_list = [
#     {'C': [0.5, 1.0, 1.5, 2.0], 'max_iter': [10000]},
    {'C': [8, 10, 12, 15], 'degree': [2, 3, 4]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45]},
    {'max_depth': [20, 25, 30], 'max_leaf_nodes': [35, 40, 45]},
]
```

```python
for i in range(1, 6):
    X = data[:, :-i]
    for clf, params in zip(clf_list, param_list):
        model = clf()
        grid_model = GridSearchCV(model, param_grid = params, cv = 3)
        grid_model.fit(X, y)
        grid_best_params['Classifier'].append(model.__class__.__name__)
        grid_best_params['Params'].append(grid_model.best_params_)

        model = clf(**grid_model.best_params_)
        score_board['Classifier'].append(model.__class__.__name__)
        score = cross_val_score(model, X, y, cv = 3)
        score_board['Accuracy'].append(score)
        score_board['Accuracy mean'].append(score.mean())
        score_board['N_features'].append(len(X[0]))
        print(model.__class__.__name__, ':: checking complete')
```

```python
score_board_df = pd.DataFrame(score_board)
score_board_df
```
