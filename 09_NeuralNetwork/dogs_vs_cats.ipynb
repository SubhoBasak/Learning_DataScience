```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
datagen = ImageDataGenerator(rotation_range = 90, width_shift_range = 0.3, height_shift_range = 0.3, shear_range = 0.2,
                            zoom_range = 0.3, horizontal_flip = True, fill_mode = 'nearest')
```

```python
img = load_img('cats_vs_dogs/cats/cats (1).jfif')
```

```python
img
```

```python
img_arr = img_to_array(img)
```

```python
img_arr
```

```python
img_arr.shape
```

```python
img_arr = img_arr.reshape((1, ) + img_arr.shape)
```

```python
img_arr.shape
```

```python
i = 0
for batch in datagen.flow(img_arr, save_to_dir = 'cats_vs_dogs/cats_preview/', save_prefix = 'cat', save_format = 'jpeg'):
    i += 1
    if i == 20:
        break
```

```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, array_to_img, load_img
```

```python
img = load_img('cats_vs_dogs/cats/cats (7).jfif')
img
```

```python
img_arr = img_to_array(img)
img_arr
```

```python
img_arr = img_arr.reshape((1, )+img_arr.shape)
```

```python
datagen = ImageDataGenerator(rotation_range = 70,
                            vertical_flip = True,
                            horizontal_flip= True,
                            width_shift_range= 0.4,
                            height_shift_range= 0.5,
                            fill_mode= 'nearest')
```

```python
i = 0
for batch in datagen.flow(img_arr, batch_size = 1, save_to_dir= 'cats_vs_dogs/cats_preview/', save_prefix = 'cat', save_format='jpeg'):
    i += 1
    if i == 20:
        break
```
