```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('BX-Books.csv', encoding = 'latin-1', error_bad_lines = False, sep = ';')
books.head()
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv', encoding = 'latin-1', sep = ';', error_bad_lines = False)
ratings.head()
```

```python
users = pd.read_csv('BX-Users.csv', encoding = 'latin-1', error_bad_lines = False, sep = ';')
users.head()
```

```python
rating_count = pd.DataFrame(ratings.groupby('ISBN')['Book-Rating'].count().reset_index().rename(columns = {'Book-Rating': 'rating_count'}))
rating_count.head()
```

```python
rating_count.sort_values('rating_count', ascending = False).merge(books, on = 'ISBN')[:10][['Book-Title', 'Book-Author', 'rating_count']]
```
