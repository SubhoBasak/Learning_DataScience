\begin{equation*}
W = \frac{Rv+Cm}{v+m}
\end{equation*}

W = weighted average<br>
R = average vote of the movie<br>
v = number of vote of the movie<br>
C = average vote in the entire report<br>
m = number of vote required for listed on top 250<br>

```python
R = df['vote_average']
v = df['vote_count']
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.7)
```

```python
df['weighted_average'] = (R*v+C*m)/(v+m)
```
