```python
df = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.rename({'movie_id': 'id'}, axis = 1, inplace = True)
```

```python
df = df.merge(credits, on = 'id')
```

```python
df.drop(['homepage', 'production_countries', 'status', 'title_x', 'title_y'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

\begin{equation*}
W = \frac{Rv+Cm}{v+m}
\end{equation*}

W = weighted average <br>
R = average for the movie as a number from 0 to 10 <br>
v = number of votes for the movie <br>
C = the mean vote across the whole report <br>
m = minimum votes required to listed in the top 250 <br>

```python
r = df['vote_average']
v = df['vote_count']
c = df['vote_average'].mean()
m = df['vote_count'].quantile(0.7)
```

```python
df['weighted_average'] = ((r*v)+(c*m))/(v+m)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(df[['weighted_average', 'popularity']])
normalized_values_df = pd.DataFrame(normalized_values, columns = ['weighted_average', 'popularity'])
normalized_values_df.head()
```

```python
df[['weighted_average', 'popularity']] = normalized_values_df
```

```python
df['score'] = df['weighted_average']*0.5+df['popularity']*0.5
```
